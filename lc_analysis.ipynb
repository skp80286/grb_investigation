```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data/mcmc_df.csv')
df1 = pd.read_csv('data/mcmc_df_trunc.csv')

swift = df[df['Filt'] == 'X-ray'].sort_values(by='Times')
swift1 = df1[df1['Filt'] == 'X-ray'].sort_values(by='Times')

plt.errorbar(swift['Times'], swift['Fluxes'], yerr=swift['FluxErrs'], marker='o', ms=10, fmt='o', capsize=3, capthick=1, alpha = 0.3)
plt.plot(swift1['Times'], swift1['Fluxes'], marker='*', color='black')
plt.xscale('log')
plt.yscale('log')
plt.minorticks_on()
plt.grid(True, which='major', alpha=0.7)
plt.grid(True, which='minor', alpha=0.3)
plt.xlabel(r'$t$ (s)')
plt.ylabel(r'$F_\nu$ (mJy)')

plt.legend(title='X-ray')
plt.show()
```
