```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def plot(filt):
    df = pd.read_csv('data/mcmc_df.csv')
    df1 = pd.read_csv('data/mcmc_df_trunc.csv')

    swift = df[df['Filt'] == filt].sort_values(by='Times')
    swift1 = df1[df1['Filt'] == filt].sort_values(by='Times')
    fig = plt.figure(figsize=(6,3))


    plt.errorbar(swift['Times'], swift['Fluxes'], yerr=swift['FluxErrs'], ms=10, fmt='o', capsize=3, capthick=1, alpha = 0.3)
    plt.plot(swift1['Times'], swift1['Fluxes'], marker='*', color='black')
    plt.xscale('log')
    plt.yscale('log')
    plt.xlim(500,5e6)
    plt.minorticks_on()
    plt.grid(True, which='major', alpha=0.7)
    plt.grid(True, which='minor', alpha=0.3)
    plt.xlabel(r'$t$ (s)')
    plt.ylabel(r'$F_\nu$ (mJy)')

    swift1np = swift1[['Times','Fluxes']].to_numpy()
    for i in range(len(swift1)-1):
        slope = np.log(swift1np[i+1,1]/swift1np[i,1])/np.log(swift1np[i+1,0]/swift1np[i,0])
        plt.text(np.sqrt(swift1np[i+1,0]*swift1np[i, 0]),
        np.sqrt(swift1np[i+1,1]*swift1np[i,1]), f"{slope:.1f}", 
                bbox=dict(boxstyle="round,pad=0.5", facecolor="white", alpha=0.5, edgecolor='none'),
                fontsize=10, fontfamily='monospace')

    plt.legend(title="'"+filt+"' band")
    plt.show()

plot('X-ray')
plot('i')
plot('r')
plot('z')
plot('g')
plot('J')
plot('VT_R')
plot('VT_B')
plot('radio(1.3GHz)')
plot('radio(6GHz)')
plot('radio(10GHz)')
```
